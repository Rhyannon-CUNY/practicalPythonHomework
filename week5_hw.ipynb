```python
def credit_rating(credit):
    '''
    This function categorizes how good a creit score is from poor to excellent
    para1 = numerical credit score
    '''
    if credit <=579:
        category ="poor"
    elif 579 < credit <= 669:
        category ="fair"
    elif 670 < credit <= 739:
        category ="good"
    elif 740 <= credit <= 799:
        category ="very good"
    else:
        category ="excellent"
    return print(f"Your credit of {credit} is {category}")
```
